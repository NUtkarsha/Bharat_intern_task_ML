```python
columns_to_plot = ['bedroom', 'gym', 'lift', 'parking', 'maintenance', 'security', 'play_area', 'clubhouse',
                   'intercom', 'garden', 'indoor_game', 'gas_connection', 'jogging_track', 'swimming_pool']

# Create count plots for each column

for column in columns_to_plot:
    plt.figure(figsize=(5, 4))
    sns.countplot(data=df, x=column)
    plt.xlabel(column)
    plt.ylabel('Count')
    plt.title(f'valuecount of {column}')
    plt.show()
```

- The countplot for the "bedroom" feature shows that the majority of houses have 2 bedrooms.
- The countplot for the "gym" feature indicates that a significant number of houses have a gym facility.
- Similarly, the countplot for the "lift" feature reveals that most houses are equipped with a lift.
- For the "parking" feature, the countplot suggests that a large proportion of houses have parking spaces.
- In contrast, the countplot for the "maintenance" feature indicates that only a few houses have maintenance services.
- The countplot for the "security" feature demonstrates that the majority of houses have security provisions.
- Similarly, the countplot for the "play_area" feature suggests that a significant number of houses have a play area.
- Interestingly, the countplot for the "clubhouse" feature shows an equal distribution of houses with and without a clubhouse.
- In the case of the "intercom" feature, there is no substantial difference between houses with and without an intercom system.
- The countplots for the "garden," "indoor_game," "gas_connection," and "jogging_track" features reveal that a small number of houses have these amenities.
- Finally, the countplot for the "swimming_pool" feature indicates a relatively balanced distribution between houses with and without a swimming pool.

```python
features_to_plot = ['area','bedroom', 'gym', 'lift', 'parking', 'maintenance', 'security', 'play_area', 'clubhouse',
                    'intercom', 'garden', 'indoor_game', 'gas_connection', 'jogging_track', 'swimming_pool']

# Create scatter plots for each feature against price
for feature in features_to_plot:
    plt.figure(figsize=(10, 6))
    sns.scatterplot(data=df, x=feature, y='price')
    plt.xlabel(feature)
    plt.ylabel('Price')
    plt.title(f'{feature} vs Price')
    plt.show()
```

```python
# distribution of location vs price
top_locations = df.groupby('location')['price'].mean().nlargest(50).index
df_top_locations = df[df['location'].isin(top_locations)]
average_price = df_top_locations.groupby('location')['price'].mean().reset_index()
average_price = average_price.sort_values(by='price', ascending=False)
plt.figure(figsize=(15, 9))
sns.barplot(data=average_price, x='location', y='price', order=average_price['location'], ci=None)
plt.xlabel('Location')
plt.ylabel('Average Price')
plt.title('Average Price for Top 50 Locations')
plt.xticks(rotation=90) 
plt.tight_layout()
plt.show()
```

- The location with the highest price is 'Worli South Mumbai', followed by 'Juhu Tara', 'Hanuman Nagar', 'Worli Sea Face', and 'Yari Road'.

```python
# Calculate the IQR
Q1 = np.percentile(df['price'], 25)
Q3 = np.percentile(df['price'], 75)
IQR = Q3 - Q1

# Define the lower and upper bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Remove outliers from the DataFrame
df = df[(df['price'] >= lower_bound) & (df['price'] <= upper_bound)]
```

```python
sns.boxplot(df['price'])
```

```python
# Calculate the IQR
Q1 = np.percentile(df['area'], 25)
Q3 = np.percentile(df['area'], 75)
IQR = Q3 - Q1

# Define the lower and upper bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Remove outliers from the DataFrame
df = df[(df['area'] >= lower_bound) & (df['area'] <= upper_bound)]
```

```python
# let's explore location column
df['location'].value_counts()
```

- We can observe that there are locations with only 1 count. To simplify the analysis, we can merge these locations with the "Other" category.

```python
location_counts = df['location'].value_counts()
locations_less_than_10 = location_counts[location_counts < 10].index.tolist()
len(locations_less_than_10)
```

```python
df.location = df.location.apply(lambda x : 'other' if x in locations_less_than_10 else x)
```

```python
df.head(20)
```

```python
dummies = pd.get_dummies(df.location)
dummies.head()
```

```python
df1 = pd.concat([df,dummies.drop('other',axis='columns')],axis='columns')
df1.head()
```

```python
df1.drop(columns='location', inplace=True)

df1.head(2)
```

```python
X = df1.drop(['price'],axis = 'columns')
y = df1.price
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score
model = LinearRegression()

model.fit(X_train, y_train)

y_pred = model.predict(X_test)

r2 = r2_score(y_test, y_pred)

print("R-squared score:", r2)
```

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor() 

rf.fit(X_train, y_train)

y_pred = rf.predict(X_test)

r_squared = r2_score(y_test, y_pred)
print("R-squared score (Random Forest):", r_squared)
```
